```python
# 기본코드 (공청기 필터1 )
test_air = [1,23,32,-5,45,0,-3,123,2,-13]
def pos(num):
    result = []
    for n in num:
        if n > 0:
            result.append(n)
    return result

print(pos(test_air))
```

```python
# 기본코드응용 (리스트컴프리헨션)
test_air = [1,23,32,-5,45,0,-3,123,2,-13]
print([n for n in test_air if n > 0])
```

```python
# filter ** 중요  filter(함수, 반복_가능한_데이터)
# 공청기2용 필터 기본코드 // 필터는 리턴이 True !! 가 중요!

test_air = [1,23,32,-5,45,0,-3,123,2,-13]
def pos_fn(n):
    return n > 0
def neg_fn(n):
    return n < 0    
print(list(filter(pos_fn ,test_air)))
print(list(filter(neg_fn ,test_air)))
```

```python
# filter 응용 (lambda ) # 위에 있는 것을 람다로
test_air = [1,23,32,-5,45,0,-3,123,2,-13]
print(list(filter(lambda x: x > 0 ,test_air)))
print(list(filter(lambda x: x < 0 ,test_air)))
```

```python
# 기본코드(가공 1)
sample = [1, 2, 3, 4, 11, 22, 33, 44]
def two_times(numberList):
    result = []
    for number in numberList:
        result.append(number*2)
    return result

result = two_times(sample)
print(result)
```

```python
# 기본코드응용 (리스트컴프리헨션)
sample = [1, 2, 3, 4, 11, 22, 33, 44]
print([n*2 for n in sample] )
```

```python
# map ** 중요 map(함수, 반복_가능한_데이터)
sample = [1, 2, 3, 4, 11, 22, 33, 44]
def two_times(x): 
    return x*2
print(list(map(two_times, sample)))
```

```python
# map 응용 (lambda)
sample = [1, 2, 3, 4, 11, 22, 33, 44]
print(list(map(lambda x : x*2 , sample)))
```

```python
# functools.reduce
def add(data):
    result = 0
    for i in data:
         result += i
    return result
data = [1,2,3,4,5]
result = add(data)
print(result)
```

```python
# operator.itemgetter()
from operator import attrgetter

class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
    def pr(self):
        print(self.name, self.age, self.grade)

students = [    Student('jane', 22, 'A'), Student('dave', 32, 'B'), Student('sally', 17, 'B'),]
result = sorted(students, key=attrgetter('age'))
result[0].pr()
result[1].pr()
result[2].pr()

# 1
print(students)
students[0].pr()
students[1].pr()
students[2].pr()
# 2
[s1, s2, s3] = students
s1.pr()
s2.pr()
s3.pr()
```

## threading

```python
import time

def long_task():  # 5초의 시간이 걸리는 함수
    for i in range(5):
        time.sleep(1)  # 1초간 대기한다.
        print("working:%s\n" % i)

print("Start")

for i in range(5):  # long_task를 5회 수행한다.
    long_task()

print("End")
```

```python
# 스레드 사용 # 한 번에 여러 가지를 한다. 단, 생각적으로 # 비동기적 처리
import time
import threading

def long_task():
    for i in range(5):
        time.sleep(1)
        print("woring:%s\n" % i)

print("Start")

threads = [threading.Thread(target=long_task) for i in range(5) ]

for t in threads:
    t.start()
    
for t in threads:
    t.join() # join으로 스레드가 종료될 때까지 기다린다.

print("End")
```

```python
# thread_test.py
import time
import threading  
start = time.time()  

def long_task():
    data = 0
    for i in range(100*1000*1000):
        # time.sleep(1)
        # print("working:%s\n" % i)
        data += i

print("Start")

threads = [threading.Thread(target=long_task) for i in range(20)]
for i in range(5):
    t = threading.Thread(target=long_task)  # 스레드를 생성한다.
    threads.append(t) 

for t in threads:
    t.start()  # 스레드를 실행한다.

print("End")
```

```python
# 멀티코어
import time
import multiprocessing
start = time.time()
def long_task():  # 5초의 시간이 걸리는 함수
    data = 0
    for i in range(100*1000*1000):
        # time.sleep(1)  # 1초간 대기한다.
        # print("working:%s\n" % i)
        data += i 
print("Start")
processes = [ multiprocessing.Process(target=long_task) for i in range(20) ]

for p in processes:
    p.start()
for p in processes:
    p.join()
print("End")
print("processes 연산시간:",time.time() - start)
```

```python
# urllib
import urllib.request

def get_wikidocs(page):
    url = 'https://wikidocs.net/{}'.format(page)
    print(url)
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36 Edg/145.0.0.0'}
    source = urllib.request.Request(url, headers=headers)
    with urllib.request.urlopen(source) as s:
        with open('wikidocs_%s.html' % page, 'wb') as f:
            f.write(s.read()) 
```

### 리스트 관련함수

```python
# append - 추가하기 *
a = []
print(a)
a.append(1)
print(a)
a.append(2)
print(a)
a.append([11,22,33])
print(a)
a.append(4)
print(a)
```

```python
# sort 리스트 정렬 *
a = [5,43,5,6,56,3,2,4,1]
b = ['gf','f','u','a','e','f','z','c']
a.sort()
b.sort()
print(a)
print(b)
```

```python
# reverse 리스트 뒤집기
a = [5,43,5,6,56,3,2,4,1]
b = ['gf','f','u','a','e','f','z','c']
print(a[::-1])
print(b[::-1])
a.reverse()
b.reverse()
print(a)
print(b)
```

```python
# sort 와 reverse 응용
a = [5,43,5,6,56,3,2,4,1]
a.sort()
a.reverse() #  a[::-1]
print(a)
b = [5,43,5,6,56,3,2,4,1]
b.sort(reverse=True) #True는 내림차순, False 오름차순이 기본값
print(b)
```

```python
# sorted 원본 유지하기
aa = [66,7,8,99,0]
print(sorted(aa))
print(sorted(aa,reverse=True))
print(aa)
```

```python
# sorted 응용하기
fruit = ['banana', 'kiwi', 'orange', 'melon', 'apple', 'strawberry']
print(sorted(fruit))
print(sorted(fruit,key=len)) # 글자 길이의 오름차순
print(sorted(fruit,key=len,reverse=True)) # 글자 길이의 내림차순
```

```python
# index 인덱스반환 *
a = [1,2,3,4,5,6,7,8,9]
print(a.index(6))
# print(a.index(11)) # 없으면 오류
```

```python
# insert 리스트에 요소삽입 *
a = [1,2,3,4,5,6]
a.insert(0,11)
print(a)
a.insert(3,22)
print(a)
a.insert(4,[1,2,3])
print(a)
a.insert(len(a),777) # 배열의 맨뒤에 추가하려면 len()사용, append와 같음
print(a)
```

```python
# remove 리스트 요소 제거
a = [1,2,3] * 3
print(a)
a.remove(2)
print(a)
a.remove(2)
print(a)
a.remove(2)
print(a)
```

```python
# pop 리스트 요소 꺼내기 **
a = [1,2,3,4,5]
out = a.pop(0) # 값을 넣으면 인덱스의 요소를 반환하고 그 요소는 삭제한다.
print(out, a)  # 값이 없을 때(기본값)는 맨 뒤에서 부터 꺼내면서 삭제.
out = a.pop(0)
print(out, a)
out = a.pop(0)
print(out, a)
out = a.pop(0)
print(out, a)
out = a.pop(0)
print(out, a)
```

```python
# count 리스트에 포함된 요소 개수세기
a = [2,3,4,6]*2 +[1,2,3]*3
print(a, a.count(4))
```

```python
# extend 리스트 확장
a = [1,2,3]
a.extend([4,5])
print(a)
```

### 변수를 만드는 여러가지 방법

```python
a, b, c = 123, 456, 789 
# (a, b) = 123, 456
# a, b = (123, 456) # 동일
print(a)
print(b)
print(c)
```

```python
[a, b, c] = ['python', 'life','cool']
print(a)
print(b)
print(c)
```

```python
a = b = c = 0
print(a)
print(b)
print(c)
```

```python
# 두 변수의 값을 바꿀 때, 파이썬만 가능
a = 33
b = 55
print(a, b)

a, b = b, a
print(a, b)
```

```python
# 다른 언어 예, temp
aa = 33
bb = 55
print(aa, bb)

temp = aa
aa = bb
bb = temp
print(aa, bb)
```

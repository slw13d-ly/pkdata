```python
import time

def outfunc(X):
    def infunc():
        start = time.time()
        print("í•¨ìˆ˜ê°€ ì‹¤í–‰ë©ë‹ˆë‹¤.")

        result = X
    
        print("í•¨ìˆ˜ ìˆ˜í–‰ì‹œê°„: %f ì´ˆ" % (time.time()-start))
        return result
    return infunc

'''=================================='''
@outfunc
def Xfunc():
    print("í•¨ìˆ˜ê°€ ì‹¤í–‰ë©ë‹ˆë‹¤.")
@outfunc
def Yfunc():
    print("í•¨ìˆ˜ê°€ ì–´ë µìŠµë‹ˆë‹¤.")
```

```python
Xfunc()
Yfunc()
```

```python
Xfunc() = outfunc(Xfunc)
Yfunc() = outfunc(Yfunc)
Xfunc()
Yfunc()
```

```python
a = outfunc(Xfunc)
b = outfunc(Xfunc)
c = outfunc(Xfunc)
a()
b()
c()
```

```python
# decorator.py
import time

def elapsed(original_func):   # ê¸°ì¡´ í•¨ìˆ˜ë¥¼ ì¸ìˆ˜ë¡œ ë°›ëŠ”ë‹¤.
    def wrapper():
        start = time.time()
        result = original_func()    # ê¸°ì¡´ í•¨ìˆ˜ë¥¼ ìˆ˜í–‰í•œë‹¤.
        end = time.time()
        print("í•¨ìˆ˜ ìˆ˜í–‰ì‹œê°„: %f ì´ˆ" % (end - start))  # ê¸°ì¡´ í•¨ìˆ˜ì˜ ìˆ˜í–‰ì‹œê°„ì„ ì¶œë ¥í•œë‹¤.
        return result  # ê¸°ì¡´ í•¨ìˆ˜ì˜ ìˆ˜í–‰ ê²°ê³¼ë¥¼ ë°˜í™˜í•œë‹¤.
    return wrapper

def myfunc():
    print("í•¨ìˆ˜ê°€ ì‹¤í–‰ë©ë‹ˆë‹¤.")

decorated_myfunc = elapsed(myfunc)
decorated_myfunc()
```

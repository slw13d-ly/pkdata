````python
# 문자열
print("%s" % '12.34 문자표현' )
# 정수
print("%10d" % 12.34 )
# 부동소수점
print("%.2f" % 12.34 )
# %%표기
print("한국에 산은 %d%% 정도 있습니다." % 70)
````

````python
# 문자열
print("%-20s입니다" % '12.34 문자표현' )
# 정수
print("%10d" % 12.34 )
# 부동소수점
print("%f" % 12.34 )
# %%표기
print("한국에 산은 %d%% 정도 있습니다." % 70)
````

````python
# 소수점 표현하기
print("%.2fkkkkk" % 3.141592)
print("%10.2fkkkkk" % 3.141592)
print("%-10.2fkkkkk" % 3.141592)
````

### format 함수를 사용한 포매팅

````python
# 숫자값을 가진 변수로 대입하기
s = "five"
ss = "I eat {0} apples".format(s)
print(ss, type(ss))
````

````python
# 2개 이상 값 넣기
s = "five"
e = 7
ss = "I eat {0} apples, so I was sic for {1} days.".format(s,e)
print(ss, type(ss))
````

````python
# 혼용해서 넣기
sss ="I ate {0} apples. so I was sick for {day} days.".format(10, day=3)
print(sss)
````

````python
# 왼쪽정렬
print("{0:<10}".format("hi"))
# 오른쪽정렬
print("{0:>10}".format("hi"))
# 가운데 정렬
print("{0:^10}".format("hi"))
````

````python
# 응용
re = "{0:.^18}".format("영수증")
print("="*20+"\n"+re+"\n"+"="*20)
````

````python
# upper - 소문자를 대문자로 바꾸기
a = 'Python is the best choice'
print(a.upper())
# lower - 대문자를 소문자로 바꾸기 
b = 'Python Is The Best Choice'
print(b.lower())
c = 'Python Is The Best Choice'
print(c[0]+c[1:].lower())
````

````python
ss = f"{'hi':^12}"
print(ss)
# lstrip - 왼쪽공백지우기
print(ss.lstrip())
# rstrip - 오른쪽공백지우기
print(ss.rstrip())
# strip - 양쪽공백지우기 *
print(ss.strip())
````

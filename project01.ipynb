```python
# 오늘의 문제 01 지역날씨
# 다음은 특정지역에 대한 날씨 API를 찾을 수 있는 링크이다.
# https://api.open-meteo.com/v1/forecast?latitude=37.5665&longitude=126.9780&current_weather=true
# 다음 주소를 힌트삼아 딕셔터리 데이터형으로 분석하고 의미있는 결과를 출력하라

import requests as req

url = "https://api.open-meteo.com/v1/forecast?latitude=37.5665&longitude=126.9780&current_weather=true"
ori = req.get(url).json()

lat = ori['latitude'] # 위도
lon = ori['longitude'] # 경도

time = ori['current_weather']['time'] # 시간
temperature = ori['current_weather']['temperature'] # 온도
windspeed = ori['current_weather']['windspeed']# 풍속
is_day = ori['current_weather']['is_day']# 낮밤
weathercode = ori['current_weather']['weathercode']# 날씨코드

result = "낮" if(is_day == 1) else "밤" # 낮/밤 구분 출력
result2 = "맑음" if(weathercode <= 2) else "흐림" # 날씨코드 구분 출력

print([ori])
print("위도:", lat, "경도:", lon)
print(f"날짜 및 시간: {time}, 온도: {temperature}도, 풍속: {windspeed}")
print("낮/밤:", result)
print("날씨:", result2)
```

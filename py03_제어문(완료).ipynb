```python
a = [int(((r*3)/2)*1.5) for r in range(0, 100)]
b = [] # [짝수만]
c = [] # [홀수만]

# 2로 나누었을 때, 나머지가 1이면 홀수 / 나누어 떨어진다면 짝수

for i in a:
    if i % 2 == 0:
        b.append(i)
    else:
        c.append(i) # pass 해도 됨
        
print(a, len(a))
print("짝수:", b, len(b))
print("홀수:", c, len(c))
```

```python
# 리스트 컴프리헨션

a = [int(((r*3)/2)*1.5) for r in range(0, 100)]
b = [i for i in a if i%2 == 0] # [짝수만]
c = [i for i in a if i%2 == 1] # [홀수만]

# 2로 나누었을 때, 나머지가 1이면 홀수 / 나누어 떨어진다면 짝수

print(a, len(a))
print("짝수:", b, len(b))
print("홀수:", c, len(c))
```

### zip 함수로 여러 리스트 함께 순회하기 **
두 개 이상의 리스트를 동시에 순회하고 싶을 때 사용.
단, 각 리스트의 길이가 같아야 함

```python
names = ['홍길동', '김철수', '이영희']
korean = [85, 92, 78]
english = [90, 88, 95]
for name, kor, eng in zip(names, korean, english):
    print(f"{name}: 국어 {kor}점, 영어 {eng}점")
```

```python
names = ['홍길동', '김철수', '이영희', '이순신']
scores = [85, 92, 78, 99]
grade = ["합격" if x > 85 else "불합격" for x in scores]  # 결과에서 거름. else 사용 시 if 문이 앞으로 가야함/ 이건 더 알아보기

for x, y, z in zip(names, scores, grade ):
    print(f"{x}님은 {y}점 이므로 {z}입니다.")
```

```python
# 주사위의 눈금 조작. 3번을 16.x% -> 25% 확률로 조작하라. 

import random as ran
n = 0
box = []
cc = 10000
choice = int(input("조작할 확률을 입력하세요.") or 10) # % 조절, 10/100 => 10%, 원래 나오는 16.x%에서 10%를 더 추가함.
my = int(input("조작할 번호를 입력하세요.") or 3)

while n < cc: # 몇 회 돌린다.
    num = ran.randint(1,6) # 1~6까지의 숫자가 랜덤으로 나옴.
    lucky = ran.randint(1,100)
    if choice >= lucky:
        box.append(my) # 10%의 my(사용자가 원하는 숫자)를 더 찍는다. 
    else:
        box.append(num) 
    n += 1
    
if cc == len(box): # 검증
    print(f"{cc}회 정상 진행 되었습니다.")
    for n in range(1, 7):
        print(f"{n}번이 {box.count(n)}번, {((box.count(n)/cc)*100):0.2f}% 등장하였습니다.")
else:
    print(f"검증 결과에 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")
```

```python
# 주사위의 눈금 조작. 3번을 16.x% -> 25% 확률로 조작하라.
# 그래프?

import random as ran
import matplotlib.pyplot as plt # 알고 있으면 조아횽

n = 0
box = []
cc = 10000
values = []

choice = int(input("조작할 확률을 입력하세요.") or 10) # % 조절, 10/100 => 10%, 원래 나오는 16.x%에서 10%를 더 추가함.
my = int(input("조작할 번호를 입력하세요.") or 3)

while n < cc: # 몇 회 돌린다.
    num = ran.randint(1,6) # 1~6까지의 숫자가 랜덤으로 나옴.
    lucky = ran.randint(1,100)
    if choice >= lucky:
        box.append(my) # 10%의 my(사용자가 원하는 숫자)를 더 찍는다. 
    else:
        box.append(num) 
    n += 1
    
if cc == len(box): # 검증
    print(f"{cc}회 정상 진행 되었습니다.")
    for n in range(1, 7):
        print(f"{n}번이 {box.count(n)}번, {((box.count(n)/cc)*100):0.2f}% 등장하였습니다.")
        values.append(box.count(n)) # 데이터 전처리기
else:
    print(f"검증 결과에 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")

color = ["skyblue", "orange", "skyblue", "orange", "skyblue", "orange"]
plt.bar(range(1, 7), values, color=color) # 순회 가능한 값의 형태로 들어가야함, 함수도 가능. (가로, 세로) / 함수안에 함수를 넣는다 => 재사용성 상승
plt.show()
```

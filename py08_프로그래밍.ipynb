```python
# abs * 절대값으로 바꾸어줌
print(abs(3))
print(abs(-3))
print(abs(-1.2))
```

```python
# all * 
# 단, all의 입력 인수가 빈 값인 경우에는 Ture를 반환
print(all([1, 2, 3, 4, 5]))
print(all([1, 2, 3, '안녕', '반가워']))
print(all([1, '', 3, 4]))
print(all([]))
```

```python
# any * 
# 단, any의 입력 인수가 빈 값인 경우에는 False를 반환
print(any([1, 2, 3, 4, 5]))
print(any([False, 2, 0, '']))
print(any(['안녕', '하세요', '!', 0]))
print(any([]))
```

```python
# 아스키코드(ASCII) - 정의: 7비트(128문자)
# 유니코드 - 정의: 전 세께 모든 문자를 표현할 수 있는 산업코드
print(chr(37))
print(chr(44033))
print(ord('a'))
print(ord('각'))
print(chr(int(hex(33)[2:])))
print(int(hex(33)[2:], 16))
```

```python
# 01) 클래스 상속받고 메서드 추가하기1 # 클래스 p.190~

class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val
        
class UpgradeCalculator(Calculator):
    def minus(self, val):
        self.value -= val
```

```python
cal = UpgradeCalculator()
cal.add(10)
cal.minus(7)

print(cal.value)
```

```python
# 02) 클래스 상속받고 메서드 추가하기2 # 클래스 p.190~

class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val

class MaxLimitCalculator(Calculator):
    def add(self, val):
        self.value += val
        if self.value > 100:
            self.value = 100
```

```python
cal = MaxLimitCalculator()
cal.add(50)
cal.add(40)

print(cal.value)
```

```python
# 14) 기록순으로 정렬하기 # operator.itemgetter p.269
from operator import itemgetter

data = [('윤서현', 15.25),
        ('김예지', 13.31),
        ('박예원', 15.34),
        ('송순자', 15.57),
        ('김시우', 15.48),
        ('배숙자', 17.9),
        ('전정웅', 13.39),
        ('김혜진', 16.63),
        ('최보람', 17.14),
        ('한지영', 14.83),
        ('이성호', 17.7),
        ('김옥순', 16.71),
        ('황민지', 17.65),
        ('김영철', 16.7),
        ('주병철', 15.67),
        ('박상현', 14.16),
        ('김영순', 14.81),
        ('오지아', 15.13),
        ('윤지은', 16.93),
        ('문재호', 16.39)]

result = sorted(data, key=itemgetter(1))
print(result)
```

```python
# 15) 청소당번 2명 뽑기 # itertools.combinations p.267
import itertools
box = ['나지혜', '성성민', '윤지현', '김정숙']
for r in itertools.combinations(box, 2):
    print(r)
```

```python
# 17) 5명에게 할일 부여하기 # itertools.zip_longest p.264
import itertools
import random

students = ['김승현', '김진호', '강춘자', '이예준', '김현주']
works = ['청소', '빨래', '설거지']

r_students = random.sample(students,len(students))

result = itertools.zip_longest(r_students, works, fillvalue='휴식')
print(list(result))
```

```python
# 18) 벽에 타일 붙이기 # 최대공약수 math.gcd p.261
import math
tile_length = math.gcd(200, 80) # 한 선 길이

w_num = 200/(tile_length)
h_num = 80/(tile_length)

print("타일 한 선의 길이:", tile_length, "필요한 타일 개수:", (w_num*h_num))
```
